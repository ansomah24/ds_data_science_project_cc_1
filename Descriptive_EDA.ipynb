TARGET VARIABLE ANALYSIS

```python
plt.figure(figsize=(12,6))
plt.hist(df_new['Price'])
plt.title("Frequency Distribution of Pric of Car")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(8,4))
plt.hist(df_new['Price'], color='red',bins=10)
plt.title("Frequency Distribution of Pric of Car")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```

```python
##using seaborn
plt.figure(figsize=(8,4))
sns.histplot(df_new['Price'], bins=30, color='blue')
plt.title("Frequency Distribution of Price of Car")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```

```python
##using 
plt.figure(figsize=(8,4))
sns.scatterplot(data=df_new, x='Age', y='Price', color='green')
plt.title("Corrrelation b/n Age and Price")
plt.xlabel("Age(Months)")
plt.ylabel("Price")
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.boxplot( x='Age', y='Price', color='green')   ### complete it
plt.title("Distribution of Fuel Type")
plt.xlabel("Fuel Type")
plt.ylabel("Price")
plt.show()
```
